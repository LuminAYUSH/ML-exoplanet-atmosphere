```python
from pycaret.regression import *
import pandas as pd
```

```python
data = np.load("/kaggle/input/exo-data/training.npy")
data2 = np.load("/kaggle/input/exo-data/testing.npy")
```

```python
pred_data = np.load("/kaggle/input/exo-data/WASP12b.npy")
new_data = pd.DataFrame(pred_data)
```

```python
data1 = data[:,:14]
df1 = pd.DataFrame(data1)
data3 = data2[:,:14]
testdata = pd.DataFrame(data3)
```

```python
s = setup(data = df1, train_size = 1.0 , test_data = testdata, target='13', session_id = 123, use_gpu=False, 
          normalize = True, normalize_method = 'minmax')
```

```python
et = create_model('et', fold=5)
```

```python
predictions_et = predict_model(et, data=new_data)
predictions_et
```

```python
rf = create_model('rf', fold=5)
```

```python
predictions_rf = predict_model(rf, data=new_data)
predictions_rf
```
