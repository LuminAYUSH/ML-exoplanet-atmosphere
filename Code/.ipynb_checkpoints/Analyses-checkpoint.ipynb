```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', size=14)
from corner import corner
from tqdm import trange
import seaborn as sns
import sklearn as sk
from sklearn import ensemble
from exobaconn import true_vs_pred, arcis_fwd2obs
from scipy.special import erf, erfinv
import pickle

def sigma2pc(sigma):
    p = erf(sigma/2**.5)
    pc_ = (1-p)/2
    pcp = 1-(1-p)/2
    return [pc_,pcp]

def pc2sigma(pc):
    p = abs(1-2*pc)
    sigma = 2**.5*erfinv(p)
    return sigma

md = pickle.load(open('../Data/metadata.p', 'rb'))
```

### Define the files with the ground truths and the posterior distributions.

The files with the ground truths are always called 'parameters' and they are, for every case, under the folder 'obs_w_noise'.

Here we show the example of the analysis of the Multinest bulk type 2 retrievals performed for simulated NIRSPEC observations.

For the 'uncomfortable' retrievals, the files are under the directory 'Uncomfortable'

```python
f_true = '../Data/Sim_Obs/NIRSPEC/type2/obs_w_noise/parameters'
method = 'Multinest' # 'Multinest' or 'CNN'
f_post = '../Data/Retrievals/'+method+'/post_nirspec_type2.p'
```

### Load those files

```python
Y = np.loadtxt(f_true)
post = pickle.load(open(f_post, 'rb'))
```

### Calculate the medians of the posteriors

```python
N = len(post)
Type = 1 if post[0].shape[1]==10 else 2
D = {1: 10, 2:7}[Type]

med_pred = np.empty((N, D))
for i in trange(N):
    med_pred[i,:] = np.median(post[i], axis=0)
```

### Calculate the distances in $\sigma$ between truths and predictions.

First, calculate the percentile of the posterior distribution in which the true value falls.

```python
pcn  = np.empty((N,D))
nsamples = np.empty(N)

for i in trange(N):
    nsamples[i] = len(post[i])
    for j in range(D):
        spost = post[i][:,j]
        pos = len(spost[spost<Y[i,j]])
        pcn[i,j] = pos/nsamples[i]
```

Then convert these percentiles into $\sigma$.

```python
names = md['names'][Type]

s = [0.5, 1.5, 2.5, 3.5]
ticks=[r'$\leq 1\sigma$', r'$(1\sigma, 2\sigma]$', r'$(2\sigma, 3\sigma]$', r'>3$\sigma$']

As = np.empty((len(pcn),D))
frac_sig = np.empty((4,D))
for i in range(D):
    t_As = pc2sigma(pcn[:,i])
    t_As[t_As==float('inf')] = 3.5
    As[:,i] = t_As
    frac_sig[:,i]=np.histogram(As[:,i], bins=[0,1,2,3,4], density=True)[0]
```

Finally, plot the histogram of those $\sigma$.

```python
fig,ax = plt.subplots(1,1, figsize=(9,6))

colors={}
colors[2] = ['C0', 'C1', 'C2', 'C6', 'C7', 'C8', 'C9']
colors[1] = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9']

hist = plt.hist(As, bins=[0,1,2,3,4], density=True, color = colors[Type])

plt.xlabel('Sigmas away from truth')
plt.xticks(s, ticks)
plt.ylabel('Fraction of predictions')

normal = abs(np.random.randn(1000*N))
sns.histplot(normal, bins=[0,1,2,3,4], stat='density', ls='--', color='grey', alpha=0.01, shrink=0.8)

from matplotlib.lines import Line2D
from matplotlib.patches import Patch, Rectangle

custom_lines = []

for i in range(D):
    custom_lines.append(Patch(facecolor=colors[Type][i], edgecolor=colors[Type][i]))
    
custom_lines.append(Patch(facecolor='white', edgecolor='k', ls='--'))

names = np.append(names, 'Ideal case')

plt.legend(custom_lines, names, ncol=2)

print('Fraction of predictions more than 3σ away from the ground truth: '+str(round(100*np.mean(hist[0][:,3]),2))+'%')
plt.title(method)
plt.ylim(0,0.8)
plt.show()
```

### Calculate the $R^2$, the bias, and plot the predicted values against the ground truths

The argument `ys` in the function `true_vs_pred` can be used to select for which parameters to plot the predicted vs true values. To do so, `ys` needs to be a list with the indices of the parameters, for example `ys=[1, 2]` would plot only the C/O ratio and the metallicity. To plot all parameters, set `ys='all'`.

The argument `dims_j` determines the number of columns in the plot.

```python
fig, R2, bias = true_vs_pred(Y, med_pred, names, a=0.35, sup_title=None, ys='all', dims_j=3)
fig.get_axes()[0].annotate(method, (0.5, 0.94), 
                            xycoords='figure fraction', ha='center', 
                            fontsize=24
                            )
print('Mean R2: ', round(np.mean(R2),2))
plt.show()
```
