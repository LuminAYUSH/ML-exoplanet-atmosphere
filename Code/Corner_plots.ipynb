```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
plt.rc('font', size=14)
from corner import corner

md = pickle.load(open('../Data/metadata.p', 'rb'))
```

### Load the posterior/s

```python
f_post1 = '../Data/Retrievals/CNN/post_nirspec_type2.p'
f_post2 = '../Data/Retrievals/Multinest/post_nirspec_type2.p'

post1 = pickle.load(open(f_post1, 'rb'))[0]
post2 = pickle.load(open(f_post2, 'rb'))[0]
```

### Define the truth

If there's no ground truth, set `truth=None`.

```python
Y = np.loadtxt('../Data/Sim_Obs/NIRSPEC/type2/obs_w_noise/parameters')
truth = Y[0]
```

###  Plot the corner plot

```python
Type = 1 if post1.shape[1]==10 else 2
CORNER_KWARGS = dict(
    smooth=.9,
    label_kwargs=dict(fontsize=16),
    levels=[0.68, 0.95],
    quantiles=[0.5],
    plot_density=False,
    plot_datapoints=True,
    fill_contours=True,
    plot_contours=True,
    show_titles=True,
    truths=truth,
    range=md['ranges'][Type]
)

# Plot the first corner plot.
fig = corner(post1, labels=md['names'][Type], color='tomato', title_kwargs=dict(fontsize=16, c='tomato',loc='left'),
              **CORNER_KWARGS)

# Get the titles of fig1. Necessary to render the titles of both fig1 and fig2.
titles=[]
for i in range(md['D'][Type]):
    axes1 = np.array(fig.axes).reshape((post2.shape[1], post2.shape[1]))
    titles.append(axes1[i,i].get_title(loc='left'))
    
# Plot the second corner plot.
weights=np.ones(len(post2))*len(post1)/len(post2)
fig2 = corner(post2, labels=md['names'][Type], fig=fig, color='royalblue', weights=weights, 
              title_kwargs=dict(fontsize=16, c='royalblue',loc='left'), **CORNER_KWARGS)

# Add a second set of titles with the medians and uncertainties of the second posterior.
axes = np.array(fig2.axes).reshape((post2.shape[1], post2.shape[1]))
for i in range(md['D'][Type]):
    ylims = axes[i,i].get_ylim()
    xlims = axes[i,i].get_xlim()
    axes[i,i].text(xlims[0], ylims[1]*1.23, titles[i], c='tomato', fontsize=16)

# Add a legend
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], marker='o', markersize=15, color='w', markerfacecolor='tomato'),
                Line2D([0], [0], marker='o', markersize=15, color='w', markerfacecolor='royalblue')]
fig2.legend(custom_lines, ['CNN', 'Multinest'])

plt.show()
```
