```python
import numpy as np
import tensorflow.keras as tfk
tfk.backend.set_floatx("float64")
import matplotlib.pyplot as plt
from exobaconn import retrieval
from tqdm import trange
import pickle
```

### Do retrievals
The function `retrieval` takes the following inputs:

`obs_file`: A file with an observation. First column must be the wavelength, second column the transit depth, third column the error in the transit depth and fourth column the spectral resolution.

`Rs`: The stellar radius in solar radii.

`instrument`: Can be either `'WFC3'` or `'NIRSPEC'`.

`Type`: Can be either `1` for free chemistry retrievals, or `2` for equilibrium chemistry retrievals.

`N`: The number of points in the posterior to sample.

`f_weights`: The file with the weights of the CNN.

`f_out`: The output file. The output will be an array of size `(N, D)`, where D is the number of parameters.

```python
Type = 2
instrument = 'WFC3'

Rs = 0.752

for i in trange(1000):
    obs = '../Data/Sim_Obs/'+instrument+'/type'+str(Type)+'/obs_w_noise/'+str(i)
    f_out = '../Data/Retrievals/CNN/'+instrument+'_type'+str(Type)+'/'+str(i)+'.dat'
    f_weights = '../Data/CNN_weights/'+instrument+'_type'+str(Type)+'.h5'

    retrieval(obs, Rs, instrument, Type, N=1000, f_weights=f_weights, f_out=f_out)
```

Put all posteriors into a dictionary and save into a pickle.

(This repository doesn't contain the individual outputs of the retrievals, but the pickled dictionaries)

```python
post = {}

direc = '../Data/Retrievals/CNN/'+instrument+'_type'+str(Type)+'/'
for i in trange(N):
    post[i] = np.loadtxt(direc + str(i) + '.dat')

pickle.dump(post, open('../Data/Retrievals/CNN/post'+instrument+'_type'+str(Type)+'.p', 'wb'))
```
